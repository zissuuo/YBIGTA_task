위에 두개는 같은 요청입니다! 뒤에 파라미터들이 ?으로 , 파라미터가 여러개일 경우 &으로 연결됩니다!

````python
import requests
import json
import time



url = 'https://api.punkapi.com/v2/beers'
headers = {'Content-Type': 'application/json'}

#페이지 1에 있는 / 2010년 1월 이전에 양조되었고 / 알콜의 양(abv)가 10 이상인 데이터 불러오기
params= {'page': 1, 'brewed_before': '01-2010', 'abv_gt': 10}

response = requests.get(url, headers=headers,params=params)

if response.status_code == 200:
    result = response.json()
    print('받아온 데이터:', result)
else:
    print('요청에 실패했습니다. 상태 코드:', response.status_code)
````

````python
import pandas as pd

df = pd.DataFrame(result)
df
````
