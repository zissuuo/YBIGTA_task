### 신입교육세션 회귀분석 과제

2. 과제

- 'Student_Performance.csv'를 이용해 다중선형회귀분석을 진행하세요!


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Student_Performance.csv', index_col = 0)
data.head()
```

2.1. 반응변수를 PI로 하고, 설명변수들을 그 외 모든 변수로 설정한 다중선형회귀분석을 진행하고 summary를 print하세요.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

Y = data.iloc[:, 4] 
X = data.iloc[:, [0, 1, 2, 3, 5]]
X = sm.add_constant(X)

model = sm.OLS(Y, X)
res = model.fit()

print(res.summary())
```

2.2. 반응변수를 PI로 하고, 설명변수들을 Study_Hours로 설정한 단순선형회귀분석을 진행하고 summary를 print하세요.

```python
form_mod = ols('PI ~ Study_Hours', data = data)
res2 = form_mod.fit()

print(res2.summary())
```

2.3. (2.1)과 (2.2)의 분석을 비교하고 Study_Hours와 PI의 관계를 짧게, 자유롭게 설명해보세요.

설명: (2.1)의 R-squared 값은 0.989, (2.2)의 R-squared 값은 0.14로 값이 상대적으로 더 큰 (2.1) 모델, 즉 다중선형회귀분석이 데이터를 더 잘 설명해준다고 말 할 수 있다.
Study Hours와 PI의 관계는, Study_Hours가 증가할 때, PI가 각 모델의 계수(2.8530 / 2.7731)만큼 증가할 가능성이 높다는 것을 의미한다. 또한, p-value가 0에 가깝게 나타났으므로 해당 계수가 유의미하다고 설명할 수 있다.

```python
X_ = X.iloc[:, 1:]

def draw_scatter(ii, jj, title):
    X_cols = np.array(X_.columns)
    fig, axs = plt.subplots(ii, jj, figsize=(10, 10), layout='constrained')
    fig.suptitle(title)
    axs_list = [(i, j) for i in range(0, ii) for j in range(0, jj)]
    for ax, col in zip(axs_list, X_cols):
        axs[ax].set_title(col)
        axs[ax].scatter(X_[col], Y)
    plt.show()

draw_scatter(2, 3, "Scatter Plot")
```
